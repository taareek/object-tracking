```python
ndim = 4
dt = 1.
```

```python
motion_mat = np.eye(2 * ndim, 2*ndim)
motion_mat
```

```python
for i in range(ndim):
    motion_mat[i, ndim+i] = dt
    print(motion_mat)
print(f"Final motion matrix:\n{motion_mat}")
```

```python
# update matrix or observation matrix
update_matrix = np.eye(ndim, 2*ndim)
print(f"The update matrix:\n {update_matrix}")
```

```python
# a sample detected object with [x,y, a, h]
# where (x,y) is current position and a,h are aspect ratio and height
measurement = np.array([53, 311, 83, 156])
mean_pos = [53, 311, 83, 156]
mean_vel = np.zeros_like(mean_pos)
mean = np.r_[mean_pos, mean_vel]
print(f"Value of mean\n: {mean}")
```

```python
measurement[3]
```

```python
std_weight_position = 1. / 20
std_weight_velocity = 1. / 160
print(f"standard deviation weight postion: {std_weight_position}")
print(f"standard deviation weight velocity: {std_weight_velocity}")
```

```python
std = [
            2 * std_weight_position * measurement[3],
            2 * std_weight_position * measurement[3],
            1e-2,
            2 * std_weight_position * measurement[3],
            10 * std_weight_velocity * measurement[3],
            10 * std_weight_velocity * measurement[3],
            1e-5,
            10 * std_weight_velocity * measurement[3]]
covariance = np.diag(np.square(std))

print(f"Shape of the covarience matrix: {covariance.shape}")
print(f"Covarience matrix:\n {covariance}")
```

```python
import torch
import torch.nn as nn
import ipdb

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)

    def forward(self, x):
        x = self.conv1(x)
        ipdb.set_trace()  # Pause execution after first convolution
        x = self.conv2(x)
        return x

net = Net()
x = torch.randn(4, 3, 128, 64)
y = net(x)  # Execution will pause inside forward()
```
